```python
countries_with_most_terrorism = data.country_txt.value_counts().head(10)
countries = list(countries_with_most_terrorism.index)
```

```python
fig, ax = plt.subplots(figsize=(14,5))
ax.bar(countries_with_most_terrorism.index,countries_with_most_terrorism.values,color='b')
plt.title('Countries with most terrorism')
plt.xlabel('Country')
plt.ylabel('Values')
plt.show()
```

```python
year = data.iyear.value_counts().head(10)
plt.figure(figsize=(13,7))
plt.bar(year.index, year.values, color='red')
plt.title('year with most number of Terrorism')
plt.xlabel('year')
plt.ylabel('Values')
plt.style.use('ggplot')
plt.show()
```

### 2014 are most terrorism years in the world has been recoreded

```python
target=data['targtype1_txt'].value_counts().head(10)
fig, ax = plt.subplots(figsize=(13,7))
ax.bar(target.index,target.values,color='r')
plt.title('Type of Targets')
plt.xlabel("Targets")
plt.ylabel("Values")
plt.style.use('ggplot')
plt.show()
```

### So we find that Private Citizens and property were targeted the most. 

```python
weapon=data['weaptype1_txt'].value_counts().head(6)
fig,drw = plt.subplots(figsize=(16,6))
drw.bar(weapon.index, weapon.values, color='g')
plt.title('mostly used weapon type')
plt.xlabel('wapon_type')
plt.ylabel('values')
plt.show()
```

### so mostly the terrorist use Explosives to attacks

```python
religious_target = data[data['targtype1_txt']=='Religious Figures/Institutions']
plt.figure(figsize=(14,6))
plt.bar(religious_target['country_txt'].value_counts().head().index, religious_target['country_txt'].value_counts().head().values,color='black')
plt.title("Countries with most terrorism on religious figures")
plt.xlabel("Country")
plt.ylabel("Values")
plt.show
```

### Countries with most terrorism on religious is Iraq

```python
plt.subplots(figsize=(18,8))
sbn.countplot(data["iyear"],order=data['iyear'].value_counts().index,palette="colorblind")
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)
plt.show()
```

```python
cities_with_most_terrorism = data.city.value_counts().head(10)
plt.figure(figsize=(16,7))
plt.bar(cities_with_most_terrorism.index,cities_with_most_terrorism.values,color='g')
plt.title('Countries with most terrorism')
plt.xlabel('Country')
plt.ylabel('Values')
plt.show()
```
